```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

## Multiples axes en un a figura

```python
fig, axes = plt.subplots()
axes.plot(x, y, "g--")
```

### ¿Podemos tener mas de una **axes** dentro de nuestra figura?

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
```

### ¿Podemos accerder a los axes atravez de indices?

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, "ro-")
```

### Los **Axes son de tipo List** dentros de nuetro lienzo(fig), asi que podemos definirnos en tuplas o listas.

> **El lienzo(fig) de tipo object**

```python
# En tuplas 
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

# En listas
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, y)
ax2.plot(y, x, "g")
```

### ¿Como acceder a axes espeficicos ?

```python
fig, axes = plt.subplots(2, 4)
axes[0,1].plot(x, y)
axes[1,3].plot(x, np.cos(y), "g^-")
```

#### ¿Como defenimos los axes en **tuplas**  y como accedemos de formas esfecifica?

```python
fig, ( (ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8) ) = plt.subplots(2, 4)
ax5.plot(x, y)
ax2.plot(x, np.cos(y), "g^-")
ax7.plot(np.tan(x), y, "y--")
```

### ¿Como podemos **personalizar** nuestro **Lienzo(fig)** ?

```python
fig, ( (ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8) ) = plt.subplots(2, 4)
ax5.plot(x, y)
ax2.plot(x, np.cos(y), "g^-")
ax7.plot(np.tan(x), y, "y--")

fig.tight_layout() # ajuste del diseño
```
