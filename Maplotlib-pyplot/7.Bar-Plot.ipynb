```python
import matplotlib.pyplot as plt
import numpy as np
```

## Â¿Como definimos variable **categoricas(_`Texto`_)** con **`Bar`** ?  
`.bar()`: Permite hacer una grafica de **barras categorica**.

- #### Definimos dos **`Listas`** una **Country(`categorica`)** y otra **Population(`numerica`)**.

```python
COUNTRY = ['MEXICO', 'COLOMBIA', 'PERU', 'BRAZIL', 'ARGENTINA']
POPULATION = [1000, 400, 800, 500, 444]
```

```python
plt.bar(COUNTRY, POPULATION)
plt.show()
```

### Â¿Como podemos **perzonalizar** nuestras grafica `.bar()`, con sus `Parameters` ?


```python
# Width: Cambia el tamaÃ±o de las barras.
plt.bar(COUNTRY, POPULATION, width=0.5)

# Color: Un Color para nuestras barras o un lista de colres.
plt.bar(COUNTRY, POPULATION, width=0.5, color=['Red', 'Blue'])




# xticks o yticks: elmentos que estan en el eje X o Y del axes.

# 1. np.arange(5): recorre en los indices de cada elmento del eje X = [0, 1, 2, 3, 4]
# 2. ("..."): Remplanzando elementos
# 3. Rotation: Angulo donde de los xticks
plt.xticks(np.arange(5),('Mexico', 'Colombia', 'Peru', 'Brazil', 'Argentina'), rotation=45) 







plt.show()
```

### `.barh()` es no mismo pero en **formato horizontal**.

```python
plt.barh(COUNTRY, POPULATION)
plt.show()
```

# Reto
- **Especificar** mÃ¡s la grafica de **arriba** ðŸ‘†, Titulo, nombre de ejes, etc.

```python
plt.figure(figsize=(12,12))
plt.bar(COUNTRY, POPULATION, width=0.5, color=["r", "g", "k", "y"])
plt.xticks(np.arange(5),('Mexico', 'Colombia', 'Peru', 'Brazil', 'Argentina'), rotation=45) 


# Tiulo
plt.title("Cantidad de Peronas por Pais")


# Nombres en los axis
plt.ylabel(ylabel="People", labelpad=0.1)
plt.xlabel(xlabel="country")

plt.grid()



plt.show()
```
