```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

### Creamos un **objecto** fig y una lista de Axes


```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.show()
```

### ¿Como podemos agregarles **Artists(diseños)** a nuestros axes con **(OO)**?

```python
# Figsize: es un parametro donde podemos ajustar le tamaño de la figura

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5)) 

# agreagando title
axes[0].set_title("Relacion X - Y")

# Defenimos  los labals en la grafica. Con el signo  " $ sin(x) $ " al inicio y final creamos una notacion matematica
axes[0].plot(x, y, label="sin(x)" ) 


# leyenda: cuando defina los labels en la grafica se mostraran con la funcion LEGEND.
axes[0].legend()


# Agregando label en los AXIS 
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")


# Agregando title a el segundo axes 
axes[1].set_title("Relacion Y - X")
axes[1].plot(y, x)

# grid: agreaga una cuadricula al axes
axes[1].grid(True)


# Agregando label en los AXIS ejes
axes[1].set_xlabel("Y")
axes[1].set_ylabel("X")


plt.show()
```

### ¿Como puedo agregar **Artists** sin la orientacion a objectos ❌ ?

```python
plt.plot(x, y, label="sin(x)")

plt.title("Titulo")

plt.xlabel("x")

plt.ylabel("y")

# Loc: Nos permite ubicar el lagend en el axes
# bbox_to_ancho: permite cambiar el tamaño en la posicion X - Y
plt.legend(loc="lower right", bbox_to_anchor=(1, 0.9))


#grid: agrega una cuadricula al axe
plt.grid(True)

plt.show()
```

> ## Nuestros **Graficos** siempre tienen que **tener un** **contexto** claro para que sean **entendidos**. 🏷️

```python
fig = plt.figure(figsize=(12,8))
plt.style.use("dark_background")


# conjunto de se sub axes.
axes = fig.add_subplot(1, 2, 1)
axes.plot(x, y)
axes.set_title("titulo")
axes.set_ylabel("y")
axes.set_xlabel("x")

# otro grafica dentro de una mismo subconjuntos.
axes.plot(x, x**2/2, color="#ddcc88", linewidth=12, linestyle="dashed", marker="o", markersize=12, markerfacecolor="#ccff00" )



axes = fig.add_subplot(2, 2, 2)
axes.plot(x, y, "r--")
axes.set_title("titulo")
axes.set_ylabel("y")
axes.set_xlabel("x")



axes = fig.add_subplot(2, 2, 4) 
axes.plot(x, y, "go-")
axes.set_title("titulo")
axes.set_ylabel("y")
axes.set_xlabel("x")
```
