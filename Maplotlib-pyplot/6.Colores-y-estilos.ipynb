```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

## ¿Como podemos ver todos los ``Styles``?

```python
print(plt.style.available)
```

## ¿Como podemos definir un `Style` ?

```python
# Style.use(" "): Nos permite usar culquier estilo. Pruebalos!!
plt.style.use("classic")
plt.style.use("seaborn-dark")
plt.style.use("dark_background")
```

```python
fig, ax = plt.subplots(figsize=(8, 8))


ax.plot(x, x+1, "r")
ax.plot(x, x+2, "y")
ax.plot(x, x+3, "g")






#show(): Solo muestra la grafica sin la instancia en memoria.
plt.show()
```

## ¿Como podemos hacer **modificaciones** de tipo `Maplab`?

```python
fig, ax = plt.subplots(figsize=(8, 8))

ax.plot(x, x+1, "r--")
ax.plot(x, x+2, "y-")
ax.plot(x, x+3, "gx-")



#show(): Solo muestra la grafica sin la instancia en memoria.
plt.show()
```

## ¿Como podemos hacer **modificaciones** de tipo `Pyplot`?

```python

fig, ax = plt.subplots(figsize=(8, 8))

# alpha: Nivel de transparencia 
ax.plot(x, x+1, color="r",  alpha=0.7)

# linewidth: Grosor de la linea
ax.plot(x, x+2,  color="#66FF66", linewidth=8, linestyle="dashed", marker="p" )


# LineStyle: Estilo de la linea.
ax.plot(x, x+3,  color="#99ff00", linestyle="--", marker="^" )

# Marker: la forma que tendran nuestra linea(p, ^, o, t)
# Markersize: El tamaño.
# Markerfacecolor: color de la figura.
ax.plot(x, x+4,  color="y", linestyle="-", marker="o", markersize=20, markerfacecolor="#ccff" )



#show(): Solo muestra la grafica sin la instancia en memoria.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


th = np.linspace(0, 2*np.pi, 128)


def demo(sty):
    
    mpl.style.use(sty)
    fig, ax = plt.subplots(figsize=(3, 3))

    ax.set_title('style: {!r}'.format(sty), color='C0')

    ax.plot(th, np.cos(th), 'C1', label='C1')
    ax.plot(th, np.sin(th), 'C2', label='C2')
    ax.legend()



# LLamdo de la funciones
for i in mpl.style.available:
    demo(i)
```
