```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

## ¿Cual es la grafica para **analizar** **distribución** y frecuencia `Histograma`?

```python
plt.hist(data)
plt.show()
```

## **Parametros**:

**`Width`:** Tamaño de las barras, **default** **= 10.**

**`histtype`:**  Estilo de la barra.

**`bins`:** Donde hay más distrubucion datos

```python
plt.hist(data, histtype="step",)
plt.show()
```

### ¿Exite **otro grafico** para **`analazir`** la **distribución** de nuestro **datos**?

```python
plt.boxplot(data)
plt.show()
```


# 🖼️ **Imagen de apoyo**  
**Maximo:** El numero más alto de nuestro datos

**Minimo:** El numero más bajo.

**Primeiro Quartil (Q1):** El `25%`de los datos esta por debajo, y el `75%` esta por encima del valor.

**Mediana Quartil (Q2):** Un `50%` de los datos es menor y el otro 50%  de los datos es mayor.

**Terceiro Quartil (Q3):** El `75%`de los datos esta por Encima, y el `25%`  esta por debajo del valor.

**Outlier:**  Que estan fuera de un rango definido.

## **Parametros**:

**`vert`:** **False** lo cambia a forma horizontal.

**`patch_artist`:**  Rellena el rango **Quartilico**.

**`notch`:** Indica donde hay mayor concentracion de los datos.

**`showfliers`:** No me muestra los **outliers( _datos fuera de rango_)**.

```python
# np.append: agrega un item a nuesto array.
data = np.append(data, 200)


plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## ¿ Hay otro **`grafico`** para la **dispercion** de **datos** ?


```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

area = (30 * np.random.rand(N)) ** 2
colors =  np.random.rand(N)
```

- Realación entre distintas variables.

```python
plt.scatter(x, y)
plt.show()
```

## **Parametros**:

**`s`:**  Otra varible.

**`c`:**  Lista de colores, o un `Array`.

**`marker`:** La forma de las figuras.

**`alpha`:** Transparencia de las figuras.

```python
plt.scatter(x, y, s=area, c=colors, alpha=0.5, marker="p")
plt.show()
```
