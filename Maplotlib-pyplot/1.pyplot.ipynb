```python
import matplotlib.pyplot as plt
import numpy as np
```

Realizamos la grafica

```python
plt.plot(x, y, "go")
#plt.plot(x, y, "r+")
```

```python
plt.plot(x, y, "rs-")
```

```python
plt.plot(x, y, "gD:")
```

```python
plt.hist(x)
```

```python
plt.pie(x)
plt.show()
```

Relacion entre dos distintas variables

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```

```python


plt.style.use('_mpl-gallery')

# make data
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# plot
fig, ax = plt.subplots()

ax.plot(x, y, color="green", linewidth=3, markersize=12)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
plt.style.use('_mpl-gallery')

# make data:
np.random.seed(3)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
