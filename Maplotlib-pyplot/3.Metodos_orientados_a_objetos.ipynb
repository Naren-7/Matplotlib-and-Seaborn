### Declaramos nuestros datos

```python
x = np.linspace( start=1, stop=5, num=11 )
y = x ** 2
```

Creamos nuestro objetos. sera nuestra base donde haremos nuestras graficas

```python
fig = plt.figure()
```

Las graficas que vamos a pintar se llaman **axes** es un estandar.

1. vamos a **agragar un axe** a nuestra figuara

```python
axes = fig.add_axes([0.1, 0.2, 0.3, 0.4,])
```

LLego el momento de graficar nuestro axes

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.2, 0.3, 0.4,])
axes.plot(x, y, "b--")
```

```python
fig = plt.figure()

# Los 2 Ultimos argumentos hacen referencia al tamaño del axes
axes = fig.add_axes([0.1, 0.2, 0.8, 0.9,]) 

# 2 Primeros argumentos hacen referencia a la posicion del axes en la figura
axes2 = fig.add_axes([0.2, 0.55, 0.4, 0.3,]) 

# Cambiando el background
axes2.set_facecolor("gray")

axes.plot(x, y, "b--")
axes2.plot(y, x, "r--")
```

# **Practicar**

Parámetros de axes: | 
-----------------| 
Pos. Eje x       |	
Pos. Eje y       |
Size graph eje x |
Size graph eje y |

```python
fig2 = plt.figure()

# add axe
axe = fig2.add_axes([0.0, 0.0, 0.3, 0.5])

# other axe
axe2 = fig2.add_axes([0.4, 0.0, 0.2, 0.5 ])


axe3 = fig2.add_axes([0.0, 0.6, 0.2, 0.5])

# other axe
axe4= fig2.add_axes([0.3, 0.6, 0.2, 0.5 ])

axe45 = fig2.add_subplot(221)
axe45.plot(x +1, y //2 )


# Graficar

axe.plot(x, y, "ro-")
axe2.plot(y, x, "go-")
axe3.plot(x, y, "bo-")
axe4.plot(y, y, "bo-")
```
