```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## **Subplot**: nos permitira **hacer mas de una grafica**

```python
x = np.linspace(0,     5,        11) 
y = x**2
```

```python
        #Filas  #columnas  #incdice
plt.subplot(1,    2,           1 )

plt.plot(x, y, "r--") # Primera grafica



plt.subplot(1, 2, 2 )

plt.hist(y) # Segunda grafica
plt.show()

```

```python
plt.subplot(1, 2, 1 )
plt.plot(x, y, "r--") # Primera grafica
plt.plot(y, x, "b:") 
 
plt.subplot(1, 2, 2 )
plt.pie(y) # Segunda grafica
plt.show()
```

```python
# fila, columna, en el que voy a graficar indice empieza desde 0
plt.subplot(3, 1, 1 ) 
plt.plot(x, y, "r--", linewidth=3) # grafico 1



# Usamos otro idice 
plt.subplot(3, 1, 2)  # grafico 2
plt.bar(x, y ) 


# Mostra Grafica
plt.show()
```

## Mi propio subconjunto de graficos


### En eje de filas

```python
plt.subplot(511) # grafica 1
plt.plot(x, y) 
plt.plot(x, y*34)


plt.subplot(512) # grafica 2
plt.pie(x)
plt.pie(y)


plt.subplot(513) # grafica 3
plt.bar(x= y, height=x)

plt.subplot(527) # grafica 4
plt.bar(x= y, height=x)

plt.show()
```
