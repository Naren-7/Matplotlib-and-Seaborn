```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Â¿Que **`graficos o opciones`** me ofrece **`Seaborn`** para trabajar con distribuciones **datos numericos**?

```python
tips = sns.load_dataset("tips")
tips.head()
```

**`Parametros`**
- **`bins`(int):** Numero de barras.

- **`cumulative`(bool):** acumula la primera barra hasta la ultima, total acumulativo.

- **`stat`(str):** Trabaja con metodos estadisticos, por **(ejemplo: 'count', 'probability', 'percent', 'frequency', 'density' )**

- **`multiple`(str):** Como quremos representar nuestros barras. por **(ejempolo: 'stack', 'layer', 'dodge', 'fill')**


```python
# bins: Numero de barras
sns.histplot(data=tips, x="total_bill", bins=15, cumulative=False, hue="sex", stat="density", multiple='dodge');
```

## Â¿ Existen mÃ¡s **graficos** para las **distribuciones**?

> ### **`kdeplot`**: Es un diagrama de **densidad**

- **`shade`(bool):** Sobrel la grafica.

- **`bw_adjust`(int):** ajusta la parametros y la grafica que estamos utilizando.

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1);
```

> ### **`ecdfplot`**: Es un diagrama **escalonado**

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count');
```

> #### **`displot`**: Analizar distribuciones. _**`displot` nos permite cambiar a distintos tipos de grafica con el parametro`kind` .**_

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist');
```
