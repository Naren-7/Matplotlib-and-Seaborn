```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

### Â¿Como podemos relacinar distiatas **variaables** de tipo **numerico**?

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size= "size");
plt.legend(loc="center", bbox_to_anchor=(1.1, 0.5 ))
```

### Â¿Como podemos **REDEFINIR MARKERS**?

```python
# Dict: guardan la forma que quiero para Lunch y Dinner.
markers = {"Lunch": "D", "Dinner": "s"}

plt.figure(figsize=(12, 8))

# Markers: NOs permite cambiar la forma de los marcadores
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size="size", markers=markers);

plt.legend(loc="center", bbox_to_anchor=(1.1, 0.5 ))
```

```python
sns.lineplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size="size", markers=markers);

plt.legend(loc="center", bbox_to_anchor=(1.1, 0.5));
```

### Â¿Que funcion nos permite **SELECIONAR OTRAS FUNCIONES PARA USAR** con el parametro **`kind`**?

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size="size", col="time");
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", style="time", size="size", kind="line");
```

```python
sns.relplot(
    
    # dataset y ejes
    data=tips,
    x="total_bill",
    y="tip",
    
    # agrupacion, columnas y tamaÃ±o.
    col="time",
    hue="time",
    size="size",
    
    
    style="sex",
    palette=["b", "r"], 
    sizes=(10, 100)
);
```
