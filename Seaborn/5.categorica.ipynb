```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset(name="tips")
tips.head(2)
```

## ¿Como Definimos graficas **categoricas** con Seaborn?

```python
sns.countplot(data=tips, x="day", hue="sex");
```

```python
plt.figure(figsize=(12, 7))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True );
```

- Es igual, pero nos muestra de manera horizontal donde más se concentran los datos.

```python
plt.figure(figsize=(12, 7))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True );
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True );
```

- **Boxplot**: funiciona para graficar varibles numericas

```python

plt.figure(figsize=(12, 7))
sns.boxplot(data=tips, x="total_bill");
```

- Combinamos graficas

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True );
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, color="0", marker="^" );
```

```python
plt.figure(figsize=(12, 7))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, palette="pastel", split=True );
```

- **violinplot**: Funciona para variables numericas.

```python
plt.figure(figsize=(12, 7))
sns.violinplot(data=tips, y="total_bill");
```

- **catplot**: maneja todos los graficos  de variable categoricas

```python

sns.catplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, kind="box", col= "smoker");
```

```python
plt.figure(figsize= (6,6))
sns.swarmplot(data = tips);
```
